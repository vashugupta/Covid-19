```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_json("https://www.mohfw.gov.in/data/datanew.json")
df.head()
```

```python
df = df.iloc[:-1:]
df.tail()
```

```python
plt.figure(figsize = (10,10))
sns.barplot(x = df['active'] , y = df['state_name'])
```

```python
data_active = df.sort_values(['active'] , ascending = False)
active = data_active.head(7)
active
```

```python
plt.figure(figsize=(5,8))
explode = [0.1,0,0,0,0,0,0]
colors = ['r','blue', 'yellow' , 'aqua' , 'green', 'm','brown']
textprops = {"fontsize" : 12}
wedgeprops = {"linewidth" : 2, "width" : 1, "edgecolor": "black"}
fontdict = {"fontsize" : 12}
plt.title("Top seven corona-affected states\nTotal active cases in India: {}\nTotal active cases in these states: {}\nTotal percentage of active cases in these states: {}%".format(sum(df["active"]), sum(active["active"]), round(sum(active["active"]) / sum(df["active"]) * 100, 4)), fontdict=fontdict)
plt.pie(active['active'] ,labels = active['state_name'], explode = explode, colors = colors, autopct="%0.2f%%" , radius=1, 
       wedgeprops = wedgeprops, textprops = textprops, pctdistance = 0.6, labeldistance = 1.1 , rotatelabels = False)
plt.legend(loc = 0)
plt.show()
```
